## Task1: Implementation of BIRCH Clustering using Python and Chateau Winery (Using given example)

### BIRCH algorithm with determined clusters

```python
# Importing rewquired libraries

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import Birch
```

```python
# Read the data

wine_data = pd.read_csv("wine_data.csv")
wine_data.head()
```

```python
wine_data.shape
```

```python
wine_data.describe()
```

```python
# Creating scatter plot of dataset

plt.figure(figsize=(10,6))
plt.scatter(wine_data["Pinot Noir"], wine_data["Champagne"])

plt.title("Exhibit 2A: The Data")

plt.xlabel("Pinot Noir")
plt.ylabel("Champagne")
```

```python
# Determining the clusters

X = wine_data.iloc[:, [1,2]].values

import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(X, method= "ward"))
plt.title("Dendogram")
plt.xlabel("Wine")
plt.ylabel("Euclidian Distance")
```

```python
# Generating a dataset using make blobs

X, clusters = make_blobs(n_samples=63, centers=3, cluster_std=.5, random_state=0)
plt.scatter(X[:,0], X[:, 1], alpha=.7, edgecolors="g")
```

```python
# Initializing and training the model

brc = Birch(branching_factor=50, n_clusters=3, threshold=1.5)
brc.fit(X)
```

```python
# Predict method

labels = brc.predict(X)
```

```python
# Plotting the data according to clusters

plt.scatter(X[:,0], X[:,1], c=labels, cmap="rainbow", alpha=.7, edgecolors="g")
```

### BIRCH Model without clusters determined

```python
# Creating a BIRCH model

model = Birch(branching_factor=50, n_clusters=None, threshold=1.5)
model.fit(X)
```

```python
predict = model.predict(X)
```

```python
# Creating Scatter plot

plt.scatter(X[:,0], X[:, 1], c = predict)
```
