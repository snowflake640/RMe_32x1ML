### Importing all necessaty python libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

### Reading the dataset

```python
df = pd.read_csv("kerala.csv")
df.head(5)   #showing top 5 data from the dataset
```

### Exploring the dataset

```python
df.info()  #shows the data type of each column, number of columns, memory usage and the number of records
```

```python
df.shape  #displays the number of records and columns
```

```python
df.describe() # summarizes the dataset's statistical properties (count, mean, min, max)
```

```python
for feature in df.columns:           # to change the object data type into categorial type
    if df[feature].dtype == 'object': 
        df[feature] = pd.Categorical(df[feature]).codes
df.corr()  # shows correlation among different variables in the dataset
```

> The closer to 1, the stronger the correlation between these variables

```python
df["FLOODS"].replace(["YES", "NO"], [1, 0], inplace=True)   #inplace here avoids creating a new dataframe
df.head(5)
```

### Feature selection

Choosing the features that have most contribution on the target output.

```python
# using selectKBest library to get the columns with strongest relationship with the target value

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

# defining x and y
X = df.iloc[:,1:14]  # all features
Y = df.iloc[:,-1]    # target output (floods)

#selecting the top 3 features:
best_features = SelectKBest(score_func = chi2, k=3)
fit = best_features.fit(X, Y)
```

```python
# creating data frames for the features and score of each feature:
df_scores = pd.DataFrame(fit.scores_)
df_columns = pd.DataFrame(X.columns)
```

```python
# combining all the features and their corresponding scores in one data frame:
features_scores = pd.concat([df_columns, df_scores], axis=1)
features_scores.columns = ["Features", "Score"]
features_scores.sort_values(by = "Score")
```

Here the top 3 features are:
<li> "SEP" </li>
<li> "JUN" </li>
<li> "JUL" </li>

### Builidng the model:


```python
# splitting the dataset into X and Y
X = df[["SEP", "JUN", "JUL"]]  # the top 3 features
Y = df[["FLOODS"]]           # target output

# splitting into dataset into the train and test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=100)

# creating a logistic regression body
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)

# predicting the result
Y_pred = logreg.predict(X_test)
print(f"test dataset{X_test}")  # test dataset
print(f"predicted result{Y_pred}")  #predicted values
```

```python
### evaluating the model's performance
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report
print("Accuracy: ", metrics.recall_score(Y_test, Y_pred, zero_division=1))
print("Precision: ", metrics.precision_score(Y_test, Y_pred, zero_division=1))
print("Cl report: ", metrics.classification_report(Y_test, Y_pred, zero_division=1))
```

so, here the model's performance is:
<li> Accuracy: 85% </li>
<li> Recall: 85% </li>
<li> Precision: 85% </li>

So, according to the model's  prediction and analysis on the model, we can reach the conclusion that the model was well enough.

This is how we build a binary logical regression model to predict the likelihood of flood in Kerala based on the provided dataset.
